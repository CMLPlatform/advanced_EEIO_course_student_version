### A short explanation on slicing

For example you have a final demand matrix (Y) with multi-index
- I only want Paddy Rice --> Y.xs("Paddy rice", level=1) 

*It gives you for every region only the product paddy rice*

- If you want Paddy rice for Austria (AT) --> Y.xs("Paddy rice", level =1).loc["AT"]
   
Another method
- Y.loc[pd.IndexSlice["AT", "Paddy Rice"]]
Here you use, the indexes of the multi-index to slice


**REMEBER**: With **iloc and loc** always first pass the rows and then the columns

```python
row = "AT"
row2 = "NL"
column = "BE"

# Selecting only countries
Y.loc[[row, row2], column]
```

```python
row = ("AT", "Paddy rice") # During the lecture we put these labels between square brackets (i.e. list), the correct way is round brackets (i.e. tupple)
column = ("BE", "Final consumption expenditure by households")

Y.loc[pd.IndexSlice[row],pd.IndexSlice[column]]
```

```python
# Using xs to slice by level 0 (i.e., region) multiindex 
Y.xs(row[0], level=0, axis=0) 
```

```python
Y.loc["AT"]
```

```python
# Slice by level 1 of the index (i.e., sector)
Y.xs("Paddy rice", level=1, axis=0)
```

```python
# slicing a region (level 0) and multiple categories (level 1)
fd_categories_columns = ["Final consumption expenditure by government", "Changes in inventories"]

Y.loc[:, pd.IndexSlice["AT",fd_categories_columns]]
```

```python
# slicing a single category in the columns 
Y.xs("Gross fixed capital formation", level=1, axis=1)
```

## Aggregate data using groupby

```python
Y_reg = Y.groupby(level=0, axis=1, sort=False).sum()
Y_reg
#level is 0, since it is first level of multi-index
#axis=1, because it is about the columns
#sort=False, because it doesn't need to sort it here
#and then you use .sum() to sum over the groups of regions
```

```python
Y_category = Y.groupby(level=1, axis=1, sort=False).sum()
Y_category
```
